```python
%%writefile inference.py
import argparse, cv2, torch, numpy as np, os, sys

# --- Add CodeFormer path manually (since it has no setup.py) ---
sys.path.append(os.path.join(os.getcwd(), 'CodeFormer'))

from gfpgan import GFPGANer

def apply_superres(frame, region, model_name):
    if model_name == 'GFPGAN':
        gfpganer = GFPGANer(
            model_path='GFPGAN/experiments/pretrained_models/GFPGANv1.4.pth',
            upscale=2,
            arch='clean',
            channel_multiplier=2,
            bg_upsampler=None)
        _, _, out = gfpganer.enhance(region, has_aligned=False, only_center_face=False, paste_back=False)
        return out

    elif model_name == 'CodeFormer':
        from basicsr.utils import img2tensor, tensor2img
        from CodeFormer.basicsr.archs.codeformer_arch import CodeFormer
        model = CodeFormer(dim_embd=512, codebook_size=1024, n_head=8, n_layers=9, connect_list=['32','64','128','256'])
        ckpt = torch.load('CodeFormer/weights/codeformer.pth')['params_ema']
        model.load_state_dict(ckpt)
        model.eval()

        with torch.no_grad():
            img = torch.from_numpy(region).permute(2,0,1).unsqueeze(0).float()/255.0
            enhanced = model(img)['output'].squeeze().permute(1,2,0).numpy()*255
        return enhanced.astype(np.uint8)

    else:
        return region


def process_video(input_path, output_path, model_name):
    cap = cv2.VideoCapture(input_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    w, h = int(cap.get(3)), int(cap.get(4))
    out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (w,h))
    print("Processing frames...")

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        h1, w1 = frame.shape[:2]
        region = frame[h1//2-60:h1//2+60, w1//2-80:w1//2+80]
        region = apply_superres(frame, region, model_name)
        frame[h1//2-60:h1//2+60, w1//2-80:w1//2+80] = cv2.resize(region, (160,120))
        out.write(frame)

    cap.release()
    out.release()
    print(f"✅ Enhanced video saved: {output_path}")


if __name__ == "__main__":
    parser = argparse.ArgumentParser()
    parser.add_argument("--input", required=True)
    parser.add_argument("--output", default="output_superres.mp4")
    parser.add_argument("--superres", choices=['GFPGAN','CodeFormer'], default='GFPGAN')
    args = parser.parse_args()
    process_video(args.input, args.output, args.superres)
```

```python
%%writefile /content/inference_fast.py
import argparse, cv2, torch, numpy as np, os, sys
from tqdm import tqdm
import warnings

warnings.filterwarnings("ignore", category=UserWarning)

# --- Add paths manually ---
sys.path.append('/content/drive/MyDrive/LatentSync_Fresh/GFPGAN')
sys.path.append('/content/drive/MyDrive/LatentSync_Fresh/CodeFormer')

from gfpgan import GFPGANer

# === Load SuperRes Models ===
def load_models(model_name):
    if model_name == 'GFPGAN':
        print("🔹 Loading GFPGAN model (fast mode)...")
        gfpganer = GFPGANer(
            model_path='/content/drive/MyDrive/LatentSync_Fresh/GFPGAN/experiments/pretrained_models/GFPGANv1.4.pth',
            upscale=2,
            arch='clean',
            channel_multiplier=2,
            bg_upsampler=None)
        return gfpganer, None

    elif model_name == 'CodeFormer':
        print("🔹 Loading CodeFormer model (fast mode)...")
        from CodeFormer.basicsr.archs.codeformer_arch import CodeFormer
        model = CodeFormer(
            dim_embd=512, codebook_size=1024, n_head=8, n_layers=9,
            connect_list=['32', '64', '128', '256']
        )
        ckpt = torch.load('/content/drive/MyDrive/LatentSync_Fresh/CodeFormer/weights/codeformer.pth',
                          map_location='cuda' if torch.cuda.is_available() else 'cpu')
        model.load_state_dict(ckpt)
        model.eval()
        if torch.cuda.is_available():
            model = model.cuda()
        return None, model

    return None, None

# === Apply SuperResolution ===
def apply_superres(region, model_name, gfpganer=None, codeformer=None):
    if region is None or region.size == 0:
        return region

    if model_name == 'GFPGAN':
        _, _, out = gfpganer.enhance(region, has_aligned=False, only_center_face=False, paste_back=False)
        return out

    elif model_name == 'CodeFormer':
        with torch.no_grad():
            img = torch.from_numpy(region).permute(2,0,1).unsqueeze(0).float()/255.0
            if torch.cuda.is_available():
                img = img.cuda()
                codeformer = codeformer.cuda()
            enhanced = codeformer(img)['output'].squeeze().permute(1,2,0).cpu().numpy()*255
        return enhanced.astype(np.uint8)

    return region

# === Process Video (Fast Mode) ===
def process_video(input_path, output_path, model_name):
    cap = cv2.VideoCapture(input_path)
    if not cap.isOpened():
        print("❌ Cannot open input video.")
        return

    fps = cap.get(cv2.CAP_PROP_FPS)
    w, h = int(cap.get(3)), int(cap.get(4))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (w, h))
    print(f"🎬 Processing {total_frames} frames using {model_name} (Fast Mode)...")

    gfpganer, codeformer = load_models(model_name)
    last_region = None

    for i in tqdm(range(total_frames)):
        ret, frame = cap.read()
        if not ret:
            break

        h1, w1 = frame.shape[:2]
        y1, y2 = max(0, h1//2 - 60), min(h1, h1//2 + 60)
        x1, x2 = max(0, w1//2 - 80), min(w1, w1//2 + 80)
        region = frame[y1:y2, x1:x2]

        # Skip empty frames safely
        if region is None or region.size == 0:
            out.write(frame)
            continue

        # Process every 3rd frame (fast mode)
        if i % 3 == 0:
            try:
                region = apply_superres(region, model_name, gfpganer, codeformer)
                last_region = region
            except Exception as e:
                print(f"⚠️ Frame {i} skipped: {e}")
                region = last_region if last_region is not None else region
        elif last_region is not None:
            region = last_region

        # Replace region safely
        if region is not None and region.size > 0:
            frame[y1:y2, x1:x2] = cv2.resize(region, (x2 - x1, y2 - y1))
        out.write(frame)

    cap.release()
    out.release()
    print(f"\n✅ Fast enhanced video saved: {output_path}")

# === Entry Point ===
if __name__ == "__main__":
    parser = argparse.ArgumentParser()
    parser.add_argument("--input", required=True)
    parser.add_argument("--output", default="output_superres_fast.mp4")
    parser.add_argument("--superres", choices=['GFPGAN', 'CodeFormer'], default='GFPGAN')
    args = parser.parse_args()
    process_video(args.input, args.output, args.superres)
```
